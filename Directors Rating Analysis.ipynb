```python
import pandas as pd
import plotly.express as px
```

Data Preparation


```python
# Read the CSV Data
df = pd.read_csv('data/movies.csv')

# Data Cleaning
#Check if there's null in Directors
print(df['Directors'].isnull().sum()) # Since Null count is 0 then we won't drop anything

#Check if there's Null or Unknown Values
print((df['Directors'] == 'Unknown').sum()) # Also Null

# Groupo Directors and Calculate Both Mocie Count and Average Rating
director_stats = df.groupby('Directors').agg(
    avg_rating=('Rating', 'mean'),
    movie_count=('Rating', 'count')
).reset_index()

# Filter the Directprs with greater than or equal 3 movie count
top_directors = director_stats[director_stats['movie_count'] >= 3].sort_values(by='avg_rating', ascending=False).head(10)
```

Data Visualization using Bar Graph

```python
fig = px.bar(
    top_directors,
    x='Directors',
    y='avg_rating',
    color='movie_count', # Changes Color Depending on Number of Movies
    title="Top 10 Directors (Min. 3 Movies) by Average Rating",
    labels={
        "Directors": " Director Name",
        "avg_rating": " Average IMDb Rating",
        "movie_count": " Total Movies"
    },
    color_continuous_scale="Viridis", # Gradient Continous scale
    template="plotly_dark" # Dark Theme
)
fig.update_yaxes(range=[0, 10]) # Manually Change the Y-axis
fig.show()
```
